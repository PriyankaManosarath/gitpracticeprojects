# Imputing missing values

```python
data_train['Item_Weight'].fillna(data_train['Item_Weight'].mean(), inplace=True)
data_train['Outlet_Size'].fillna(data_train['Outlet_Size'].mode()[0], inplace=True)
```

```python
data_train['Outlet_Identifier'].unique()
```

2.First Two characters in "Item Type" Combined have a catch to find whether it is food or NON consumable or Drink, 
with this in information, we will create a new column Item_Type_Code

```python


# First Two characters in Item Type Combined have a catch to find whether it is food or NON consumable or Drink, with this in
#Get the first two characters of ID:
data_train['Item_Type_Code'] = data_train['Item_Identifier'].apply(lambda x: x[0:2])
#Rename them to more intuitive categories:
data_train['Item_Type_Code'] = data_train['Item_Type_Code'].map({'FD':'Food',
                                                             'NC':'Non-Consumable',
                                                             'DR':'Drinks'})
data_train['Item_Type_Code'].value_counts()

#Dropping Item_Type as it is categorized to different types
data_train.drop(['Item_Type'],axis=1,inplace=True)
data_train.drop(['Item_Identifier'],axis=1,inplace=True)
```

```python
#Correcting data

#Replacing value in Item_Fat_Content with "Low Fat" and "Regular"
data_train['Item_Fat_Content'] = data_train['Item_Fat_Content'].replace({'LF':'Low Fat',
                                                             'reg':'Regular',
                                                             'low fat':'Low Fat'})
data_train['Item_Fat_Content'].value_counts()
```

```python
data_train.head()
#data_train.dtypes
```

```python
#Getting dummies using one hot encoding
data_train = pd.get_dummies(data_train, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type',
                             'Item_Type_Code','Outlet_Identifier'], drop_first=True)

```

```python
data_train.to_csv("data_train_cleaned.csv",index=False)
```

#Defining the architecture of the model

```python
import keras
import tensorflow as tf
```

```python
# importing the sequential model
from keras.models import Sequential
```

```python
# importing different layers from keras
from keras.layers import InputLayer, Dense 
```

```python
input_neurons= data_train_x.shape[1]
```

```python
input_neurons
```

```python
output_neurons = 1
```

```python
number_of_hidden_layers = 3
neuron_hidden_layer_1 = 15
neuron_hidden_layer_2 = 10
neuron_hidden_layer_3 = 5
```

```python
# defining the architecture of the model
model = Sequential()
model.add(InputLayer(input_shape=(input_neurons,)))
model.add(Dense(units=neuron_hidden_layer_1, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_2, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_3, activation='relu'))
model.add(Dense(units=output_neurons, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_absolute_error',optimizer='Adam',metrics=['accuracy'])
```

```python
model_fit_train=model.fit(data_train_x,data_train_y,epochs=500)
```

```python
prediction = model.predict(data_train_x)
```

```python
mse=metrics.mean_squared_error(prediction,data_train_y)
```
